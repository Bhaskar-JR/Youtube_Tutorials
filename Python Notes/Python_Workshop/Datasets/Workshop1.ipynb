# If Clause
---

````python
f = 1
n = input("Enter a value for n:")
if isinstance(n, int): # Checks whether the value is integer or not
    if n is 1 or 0:
        print "Execution at if"
        print "The value of {}! = {}".format(n, f)
        exit
    elif n<0:
        print "Execution at elif"
        print "n is negative!"
    else:
        print "Execution at else"
        for i in range(2, n+1):
            f *= i
        print "The value of {}! = {}".format(n, f)
else:
    print "Factorial does not exist"
````

# Functions
---
Functions are data type independent. The data type issues come only when the code has a mis-match.
If you have a code that is data type independent, then functions can work very easily.

````python
def fact(n):
    f = 1 # This is a local variable. Accessible only inside the function
    if isinstance(n, int): # Checks whether the value is integer or not
        if n is 1 or 0:
            print "Execution at if"
            print "The value of {}! = {}".format(n, f)
            exit
        elif n<0:
            f = 0
            print "Execution at elif"
            print "n is negative!"
        else:
            print "Execution at else"
            for i in range(2, n+1):
                f *= i
            print "The value of {}! = {}".format(n, f)
    else:
        f = 0
        print "Factorial does not exist"
    return f
n = input("Enter a value for n:")
f = fact(n)
print f
````

# Importing Data from text files
---

* You need a file pointer to open and close file
* Collect all data as a single massive string
* Split the lines
* Break lines using separators and assign them to variables.

````python
# Show the file before splitting
# This is done using Lists
f = open('rain_data.txt','r') # File name, access type. # r - read, w - write, rw - read+write
contents = f.readlines()
y = []
sec = []
rain = []
for line in contents:
    a, b, c = line.split()
    y.append(a)
    sec.append(b)
    rain.append(c)
````

## Using Numpy Arrays
---

The above can be done using numpy arrays as well

````python
# This is done using numpy arrays
import numpy as np
f = open('rain_data.txt','r') # File name, access type. # r - read, w - write, rw - read+write
contents = f.readlines()
y = np.zeros(30597,dtype=np.int16)
sec = np.zeros(30597)
rain = np.zeros(30597)
i = 0
for line in contents:
    y[i], sec[i], rain[i] = line.split()
    i += 1
print sec
print np.shape(sec)
f.close()
````

# Histograms
---

We need to know a few things before that! Let's calculate random numbers!

````python
import matplotlib.mlab as mlab 
# Numerical Python functions written for compatability with MATLAB commands with the same names.

m , s = 100 , 15
iq = np.random.normal(m, s, 1000) 
# Setting a random Human IQ data, normally distributed with mean = 100, sd = 15

num_bins = 50 # Number of groups we want
n, bins, patches = plt.hist(iq, num_bins, normed=1, facecolor='green')
# n = counts in each group
# bins = array of edges. (n groups have n+1 edges)
# patches = other variables for making the histogram plot (not needed)
# normed = 1 or True, then it returns n as fraction. Else, returns n as plain numbers
# Sum of fractions = 1 and, sum of counts = total number of entries
l = mlab.normpdf(bins, m, s) # Creates the pdf of normal distribution
plt.plot(bins, l, 'r', label='pdf')
plt.xlabel('IQ bins')
plt.ylabel('Probability')
plt.title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')
plt.axis()
plt.legend()
plt.grid('on')
plt.savefig('IQ_hist.png',format='png',dpi=200)
````

# Plot Animation
---

This will be really useful for observing the data! 

How to do?

* The idea is to set a loop running between a finite range. 
* Plot the values in each loop and clear them up before plotting the next loop's value. 
* Set a pause timer to delay the computation time.

````python
% matplotlib qt
# The above command is to be used only when running in an ipython terminal
# Not needed while running generally
for i in range(0, 31):                                 # Start a loop!
    plt.clf()                                          # Clear the figure
    us = np.linspace(-20, 20, 11, endpoint=True)       # For manual adjustment of contours
    plt.contourf(lon,lat,u[i][:][:],us,cmap='seismic') # Contour plotting the values
    plt.xlabel('Longitude', fontsize=10)               # X axis label
    plt.ylabel('Latitude', fontsize=10)                # Y axis label
    plt.xticks(np.arange(0,360,30))                    # X axis grid lines
    plt.yticks(np.arange(-90,100,15))                  # Y axis grid lines
    plt.title('Zonal Velocity in $ms^{-1}$ at t = '+str(i))
    plt.suptitle("Netcdf Exercise!")                   # Super title
    plt.grid(True)                                     # Enabling gridding
    plt.axis((0, 360, -90, 90))                        # Making axis rigid
    plt.colorbar(ticks=us)                             # Enabling colour bar with user defined breaks
    matplotlib.rcParams['contour.negative_linestyle'] = 'solid'
    plt.contour(lon, lat, u[i][:][:], us, colors='k')  # For plotting contours
    plt.pause(0.001)                                   # To delay the process.
print "Do this and see what happens!"
````

## Animation
---

What does it need?

* A dummy figure
* A rigid axis
* matplotlib.animation
* An animation function

````python
import numpy as np
from netCDF4 import Dataset as dt
import matplotlib
from matplotlib import pyplot as plt    
from matplotlib import animation as anim  # This is vital!

filestr = 'ncfile2.nc' # This is the file name
ncfile = dt(filestr, 'r') # Create a netcdf object in read mode
lat = np.array(ncfile.variables['latitude'][:],dtype=np.float32)

lon = np.array(ncfile.variables['longitude'][:],dtype=np.float32)
u = np.array(ncfile.variables['u10'][:],dtype=np.float32)
us = np.linspace(-20, 20, 11, endpoint=True)

fig = plt.figure()                        # Create a dummy figure
ax = plt.axes([0, 360, -90, 90])          # Set the axis rigid
matplotlib.rcParams['contour.negative_linestyle'] = 'solid'

# This is an initialization function
#def init():
#    cont.set_data([],[],[])
#    return cont,

# This is the animation function
def animate(i): 
    plt.clf()
    #print i
    cont = plt.contourf(lon, lat, u[i][:][:], us, cmap='seismic')
    plt.title('Zonal Velocity in $ms^{-1}$ at t = '+str(i))
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.xticks(np.arange(0,360,30))                   
    plt.yticks(np.arange(-90,100,15)) 
    plt.suptitle("Netcdf Exercise")
    plt.grid(True)
    plt.colorbar(ticks=us)
    plt.contour(lon, lat, u[i][:][:], us, colors='k')  # For plotting contours
    #plt.pause(0.001) # Do not do this while animating! Memory Overload!
    return cont,    # Return the handle

ani = anim.FuncAnimation(fig, animate, interval = 1000, frames=31)  # Invoke the animation function.
# Pass the animate function as an argument.
# Interval sets the time period between the frames in milli-second
# fig is the figure handle
# init_func is the initialization function
# Set the number of frames needed. This data set has 31 time snaps!
# Hence, set frames = 31

ani.save('animation.mp4',fps=1)
# To save the video
# fps = frames per second. Choose a value based on frames on top!
````

# Interpolation

````python
% matplotlib inline
import numpy as np
from scipy import interpolate

mat1 = {}

x = np.arange(0.0, 10.0)
y = np.exp(-x/3.0)
f = interpolate.interp1d(x,y) # Set an interpolation function

x1 = np.linspace(0, 9.0, 100)
y1 = f(x1)                    # Interpolated variable for y

mat1['x'] = x
mat1['y'] = y

plt.plot(x, y, 'o', x1, y1, '-')
````

````python
% matplotlib inline
x = np.arange(-5.01, 5.01, 0.25)
y = np.arange(-5.01, 5.01, 0.25)

xx, yy = np.meshgrid(x, y) # Generates a 2-D mesh

z = xx**2 + yy**2
f = interpolate.interp2d(xx, yy, z, kind='cubic')

x1 = np.arange(-5.01, 5.01, 1e-2)
y1 = np.arange(-5.01, 5.01, 1e-2)

z1 = f(x1, y1)

mat1['x1'] = x1
mat1['y1'] = y1
mat1['z1'] = z1

plt.contourf(x, y, z)
````

# Re-gridding Data
---
Applicable for 2D data

````python
import numpy as np
# Step 0: Import interpolate option
from scipy import interpolate as inter

# Step 1: Set a value to be regridded along with mesh

x = np.linspace(-4, 4, 9)
y = np.linspace(-4, 4, 9)
x1, y1 = np.meshgrid(x, y)

z1 = x1**2 + y1**2

# Step 2: Create a new range for the regridding along with mesh

xi = np.linspace(-4, 4, 17)
yi = np.linspace(-4, 4, 17)
x2, y2, = np.meshgrid(xi, yi)

# Step 3: Create a temporary variable
tck = inter.bisplrep(x1, y1, z1, s=1)  
# Bivariate B-spline representation of surface
# u is a smoothing factor

# Step 4: Interpolate the values using the variable above
z2 = inter.bisplev(xi, yi, tck)

import matplotlib.pyplot as plt
plt.subplot(211)
plt.contourf(x1, y1, z1)
plt.title('z without regridding')
plt.subplot(212)
plt.contourf(x2, y2, z2)
plt.title('z after regridding')
plt.show()
````

Do you see any difference? Regridding takes some time. So, it is better your memory is less loaded before attempting this.
