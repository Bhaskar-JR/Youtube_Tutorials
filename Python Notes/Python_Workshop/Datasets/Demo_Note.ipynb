````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = np.linspace(0, 4*np.pi, 2001)
````

````python
x
````

````python
y = np.sin(x)
````

````python
# To get the plot in a new window, use magic command matplotlib qt or matplotlib Tkinter
# To get the plot embedded, use magic command matplotlib inline
% matplotlib inline
````

````python
plt.plot(x, y, 'r')
````

````python
% matplotlib qt
plt.plot(x, y, 'r')
````

````python
% matplotlib inline
plt.plot(x, y, 'r', label='Sine of X')
plt.legend()
plt.xlabel("Angle in Radians")
plt.ylabel("Value")
plt.axis([0, 13, -1.2, 1.2])
plt.xticks(np.linspace(0, 13, 14))
plt.yticks(np.linspace(-1.2, 1.2, 9))
plt.grid()
````

````python
z = x**2 + y
````

````python
plt.plot(x, z)
````
